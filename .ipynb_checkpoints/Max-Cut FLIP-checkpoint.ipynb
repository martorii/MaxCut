# Local Max-k-cut solved with FLIP
<h3>Import libraries</h3>
   

```python
import numpy as np
import networkx as nx
import networkx.drawing
import matplotlib.pyplot as plt
import time

%matplotlib
```

<h3>Define main functions</h3>

```python
def PlotGraph(G, partition = None):
    G = nx.from_numpy_matrix(W)
    pos = nx.spring_layout(G)
    # Draw the graph according to node positions
    if partition is not None:
        colors= np.fromiter(partition.values(), dtype=int)+1
        nx.draw(G, pos, with_labels=True, node_color=colors)
    else:
        nx.draw(G, pos, with_labels=True)
    # Create edge labels
    labels = nx.get_edge_attributes(G, 'weight')

    # Draw edge labels according to node positions
    nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

    plt.show()
    return

def InitializeGraph(n, min_weight, max_weight):
    W = np.random.uniform(low=min_weight, high=max_weight, size=(n,n))
    W = np.round((W + W.T)/2,2)
    np.fill_diagonal(W, 0)
    return W

def GetKRandomPartition(n, k):
    partition = {}
    for i in range(n):
        partition[i]=np.random.randint(low=0, high=k, size=1)[0]
    return partition

def CutCost(W, partition, n, k):
    z = 0
    identity = np.identity(k)
    for i in range(n):
        for j in range(i, n):
            z += (1-identity[partition[i], partition[j]])*W[i,j]
    return z
        
def CutGainAfterFlip(W, initial_partition, final_partition, n):
    # Vi: initial partition of the vertex v
    # Vj: final partition of the vertex v
    # z: current cut
    weights_to_add = 0
    weights_to_substract = 0
    
    diff_partition = {key: initial_partition[key] - final_partition.get(key, 0) for key in initial_partition}
    
    # v: vertex that flipped
    v = [key for key, val in diff_partition.items() if val!=0][0]
    Vi = initial_partition[v]
    Vj = final_partition[v]
    
    vertices_in_Vi = [key  for (key, value) in initial_partition.items() if value == Vi]
    vertices_in_Vj = [key  for (key, value) in initial_partition.items() if value == Vj]
    for vi in vertices_in_Vi:
        weights_to_add += W[v, vi]
    for vj in vertices_in_Vj:
        weights_to_substract += W[v, vj]
    z = weights_to_add - weights_to_substract
    return z
        
    
def GreedyBestFlip(W, partition, n, k, z):
    cut_costs = []
    partitions_flip = []
    # Permutate vertices so that we do not always start by the same one
    order = np.random.permutation(range(n))
    for i in order:
        for j in range(1, k):            
            new_partition = partition.copy()
            new_partition[i] = (new_partition[i]+j)%k
            partitions_flip.append(new_partition)
            cut_costs.append(CutGainAfterFlip(W, partition, new_partition, n))
      
    # Convert the list to a numpy array
    cut_costs = np.asarray(cut_costs)
    if np.any(cut_costs[cut_costs>0]):
        # New best partition was found
        best_index = np.argmax(cut_costs)
        best_partition = partitions_flip[best_index]
        new_cut_cost = z + cut_costs[best_index]
        return best_partition, new_cut_cost
    else:
        # We are in a local optimum
        return partition, z
    
def FirstBetterFlip(W, partition, n, k, z):
    # Permutate vertices so that we do not always start by the same one
    order = np.random.permutation(range(n))
    for i in order:
        for j in range(1, k):            
            new_partition = partition.copy()
            new_partition[i] = (new_partition[i]+j)%k
            new_z = CutGainAfterFlip(W, partition, new_partition, n)
            if new_z > 0:
                return new_partition, z + new_z
      
    return partition, z
    
def SolveMaxCut(W, n, k, z, initial_partition, method):
    tol = 1e-05
    maxIter = 10000000
    it = 0
    
    zt1 = -1
    zt2 = CutCost(W, initial_partition, n, k)
    
    # In case the cut is already optimal
    new_z = zt2
    
    partition = initial_partition
    
    start = time.time()
    if method == "GreedyBestFlip":
        while(abs(zt1-zt2)>tol and it<maxIter):
            partition, new_z = GreedyBestFlip(W, partition, n, k, zt2)
            zt1 = zt2
            zt2 = new_z
            it = it + 1
    elif method == "FirstBetterFlip":
        while(abs(zt1-zt2)>tol and it<maxIter):
            partition, new_z = FirstBetterFlip(W, partition, n, k, zt2)
            zt1 = zt2
            zt2 = new_z
            it = it + 1
            
    end = time.time()
    elapsed_time = np.round(end - start, 4)
    return partition, new_z, elapsed_time, it
```

<h2>Create one instance of the problem</h2>

<h3>Define main parameters</h3>

```python
# Number of nodes
n = 6
# Number of partitions (k-cut)
k = 3
# Define min weight and max weight of the graph
min_weight=-1
max_weight=1

# Create graph and initial partition
W = InitializeGraph(n, min_weight, max_weight)

initial_partition = GetKRandomPartition(n, k)
initial_z = CutCost(W, initial_partition, n, k)

PlotGraph(W, initial_partition)
```

<h3>Solve local Max-k-cut instance</h3>

<h4>Greedy Best Flip</h4>

```python
partition, z, elapsed_time, steps = SolveMaxCut(W, n, k, initial_z, initial_partition, "GreedyBestFlip")
print('Greedy Best Flip')
print('Initial cut: ' + str(initial_z))
print('Last cut: ' + str(z))
print('Number of steps: ' + str(steps))
print("Elapsed time: " + str(elapsed_time) + " s.\n")
PlotGraph(W, partition)
```

<h4>First Better Flip</h4>

```python
partition, z, elapsed_time, steps = SolveMaxCut(W, n, k, initial_z, initial_partition, "FirstBetterFlip")
print("First Better Flip")
print('Initial cut: ' + str(initial_z))
print('Last cut: ' + str(z))
print('Number of steps: ' + str(steps))
print("Elapsed time: " + str(elapsed_time) + " s.")
PlotGraph(W, partition)
```

<h2>Solve a grid of instances - Average complexity</h2>

<h3>Define main function</h3>

```python
def RunGridMaxCutAverage(min_nodes, max_nodes, step_nodes, min_k, max_k, min_weight, max_weight, iters_for_nk, method):

    # Create nodes grid
    # nodes = np.linspace(min_nodes, max_nodes, num=step_nodes, dtype=int)
    nodes = np.logspace(np.log(min_nodes), np.log(max_nodes), num=step_nodes, base=np.exp(1))
    # Remove decimal part
    nodes = np.floor(nodes)
    # Convert to integer
    nodes = nodes.astype(int)
    
    ks = range(min_k, max_k+1)

    times_mean = np.zeros((len(nodes), len(ks)))
    gaps_mean = np.zeros((len(nodes), len(ks)))
    steps_mean = np.zeros((len(nodes), len(ks)))
    steps_sd = np.zeros((len(nodes), len(ks)))
    
    i = 0
    j = 0

    for ni in nodes:
        for ki in ks:
            # times = 0
            # gaps = 0
            steps = np.zeros(iters_for_nk)
            for it in range(1, iters_for_nk):
                # Create graph and initial partition
                W = InitializeGraph(ni, min_weight, max_weight)
                initial_partition = GetKRandomPartition(ni, ki)
                # Get initial cost value
                initial_z = CutCost(W, initial_partition, ni, ki)
                # Get next local maximum
                if method == "FirstBetterFlip":
                    partition, z, elapsed_time, n_steps = SolveMaxCut(W, ni, ki, initial_z, initial_partition, "FirstBetterFlip")
                elif method == "GreedyBestFlip":
                    partition, z, elapsed_time, n_steps = SolveMaxCut(W, ni, ki, initial_z, initial_partition, "FirstBetterFlip")
                # Save results
                # times += elapsed_time
                steps[it]=n_steps
                # Think of another measure of the gap
                # gaps += abs((z-initial_z)/initial_z)

            steps_mean[i,j] = np.mean(steps)
            steps_sd[i,j] = np.std(steps)/np.sqrt(iters_for_nk)
            # gaps_mean[i,j] = np.mean(gaps)
            j = j + 1
        j = 0
        i = i + 1
    return steps_mean, steps_sd, nodes
```

<h3>Define grid parameters</h3>

```python
# Nodes
min_nodes = 5
max_nodes = 500
num_nodes = 50

# K-cuts
min_k=2
max_k=2

# Number of instances for each combination
iters_for_nk=10

# Method
method = "FirstBetterFlip"
```

<h3>Run FLIP</h3>

```python
# Go
steps_mean, steps_sd, nodes = RunGridMaxCutAverage(
    min_nodes, max_nodes, num_nodes, min_k, max_k, min_weight, max_weight, iters_for_nk, method)
```

<h3>Show results</h3>

```python
# Show only results for first k
run_steps = steps_mean[:,0]
run_steps_sd = steps_sd[:, 0]
x = nodes.reshape(-1, 1)

# Visualising results 
plt.scatter(x, run_steps, color = 'blue') 
plt.title('Steps vs. Nodes') 
plt.xlabel('Nodes') 
plt.ylabel('Run steps')
plt.errorbar(x,run_steps,yerr=run_steps_sd, linestyle="None")
plt.show()
```

<h3>Assume polynomial</h3>

```python
# Show results
x = np.log(nodes.reshape(-1, 1))
y_exp = np.log(run_steps)
plt.scatter(x, y_exp, color = 'blue')

# Linear regression
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.mstats.linregress(x,y_exp)

# Linear regression prediction
y_reg = intercept + slope*x
plt.plot(x, y_reg, color = 'red', label='y=' + str(np.round(np.exp(intercept), 2)) + 'x^' + str(np.round(slope, 2)))
plt.title('Polynomial behavior: r=' + str(np.round(r_value, 3))) 
plt.xlabel('ln(Nodes)') 
plt.ylabel('ln(Steps)')
# Error propagation: d(ln(x))=d(x)/x
plt.errorbar(x,y_exp,yerr=run_steps_sd/run_steps, linestyle="None")
plt.legend()
plt.show()
```

<h3>Assume exponential</h3>

```python
# Show results
x = nodes.reshape(-1, 1)
y_exp = np.log(run_steps)
plt.scatter(x, y_exp, color = 'blue')

# Linear regression
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.mstats.linregress(x,y_exp)

# Linear regression prediction
y_reg = intercept + slope*x
plt.plot(x, y_reg, color = 'red', label='y=' + str(np.round(np.exp(intercept), 2)) 
         + '*(' + str(np.round(np.exp(slope), 2)) + ')^x')
plt.title('Exponential behavior: r='+ str(np.round(r_value, 3))) 
plt.xlabel('ln(Nodes)') 
plt.ylabel('ln(Steps)')
plt.errorbar(x,y_exp,yerr=run_steps_sd/run_steps, linestyle="None")
plt.legend()
plt.show()
```

<h2>Solve a grid of instances - Smoothed complexity</h2>

```python
def RunGridMaxCutSmoothed(min_nodes, max_nodes, step_nodes, min_k, max_k, 
                                  min_weight, max_weight, iters_for_nk, iters_for_w, method, sigma):

    # Create nodes grid
    # nodes = np.linspace(min_nodes, max_nodes, num=step_nodes, dtype=int)
    nodes = np.logspace(np.log(min_nodes), np.log(max_nodes), num=step_nodes, base=np.exp(1))
    # Remove decimal part
    nodes = np.floor(nodes)
    # Convert to integer
    nodes = nodes.astype(int)
    
    ks = range(min_k, max_k+1)

    steps_mean_max = np.zeros((len(nodes), len(ks)))
    step_means = np.zeros(iters_for_w)
    
    i = 0
    j = 0
    
    for ni in nodes:
        for ki in ks:
            # times = 0
            # gaps = 0
            for w in range(1, iters_for_w):
                W = InitializeGraph(ni, min_weight, max_weight)
                steps = np.zeros(iters_for_nk)
                for it in range(1, iters_for_nk):
                    # Create initial partition
                    W_k_it = W + np.random.normal(loc=0, scale=sigma, size=(ni,ni))
                    initial_partition = GetKRandomPartition(ni, ki)
                    # Get initial cost value
                    initial_z = CutCost(W_k_it, initial_partition, ni, ki)
                    # Get next local maximum
                    if method == "FirstBetterFlip":
                        partition, z, elapsed_time, n_steps = SolveMaxCut(W_k_it, ni, ki, initial_z, initial_partition, "FirstBetterFlip")
                    elif method == "GreedyBestFlip":
                        partition, z, elapsed_time, n_steps = SolveMaxCut(W_k_it, ni, ki, initial_z, initial_partition, "FirstBetterFlip")
                    # Save results
                    # times += elapsed_time
                    steps[it]=n_steps
                    # Think of another measure of the gap
                    # gaps += abs((z-initial_z)/initial_z)
                step_means[w] = np.mean(steps)

            steps_mean_max[i,j] = np.max(step_means)
            # gaps_mean[i,j] = np.mean(gaps)
            j = j + 1
        j = 0
        i = i + 1
    return steps_mean_max, nodes
```

```python
# Nodes
min_nodes = 3
max_nodes = 10
num_nodes = 4

# K-cuts
min_k=2
max_k=2

# Number of instances for each combination
iters_for_nk=3

# Method
method = "FirstBetterFlip"

# Weight distributions
w = 10

# sigma
sigma = 0.1
```

```python
# Go
steps_mean_max, nodes = RunGridMaxCutSmoothed(min_nodes, max_nodes, num_nodes, min_k, max_k, min_weight, max_weight, iters_for_nk, w, method, sigma)
```

<h3>Assume polynomial behavior</h3>

```python
# Show results
x = np.log(nodes.reshape(-1, 1))
y_exp = np.log(steps_mean_max)
plt.scatter(x, y_exp, color = 'blue')

# Linear regression
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.mstats.linregress(x,y_exp)

# Linear regression prediction
y_reg = intercept + slope*x
plt.plot(x, y_reg, color = 'red', label='y=' + str(np.round(np.exp(intercept), 2)) + 'x^' + str(np.round(slope, 2)))
plt.title('Polynomial behavior: r=' + str(np.round(r_value, 3))) 
plt.xlabel('ln(Nodes)') 
plt.ylabel('ln(Steps)')

plt.legend()
plt.show()
```

<h3>Assume exponential behavior</h3>

```python
# Show results
x = nodes.reshape(-1, 1)
y_exp = np.log(steps_mean_max)
plt.scatter(x, y_exp, color = 'blue')

# Linear regression
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.mstats.linregress(x,y_exp)

# Linear regression prediction
y_reg = intercept + slope*x
plt.plot(x, y_reg, color = 'red', label='y=' + str(np.round(np.exp(intercept), 2)) 
         + '*(' + str(np.round(np.exp(slope), 2)) + ')^x')
plt.title('Exponential behavior: r='+ str(np.round(r_value, 3))) 
plt.xlabel('Nodes') 
plt.ylabel('ln(Steps)')
plt.legend()
plt.show()
```
